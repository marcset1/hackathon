# pourcentages interessants a etudier

# pourcentages de lignes vides pour la colonne poids

```python
per_poids = (sheet1['Poids'].isnull().sum() / len(sheet1)) * 100
per_poids
```

```python
import matplotlib.pyplot as plt

plt.pie([per_poids, 100 - per_poids], labels=['moins sensibilise', 'mieux averti'], autopct='%1.1f%%')
poids = sheet1['Poids']
plt.title(f'pourcentage de {poids.name} non renseignes')
plt.show()
```

<h3>pourcentage d'etudiants ayant renseignes leur poinds</h3>

```python
#recuperer l'ensemble des etudiants
students = sheet1[sheet1['Profession'].str.lower() == 'etudiant']
students
```

```python
per_poids_students = (len(students[students['Poids'].notnull()]) / len(students)) * 100
per_poids_students
```

```python
plt.pie([per_poids_students, 100-per_poids_students], labels=['Etudiants avec poids', 'Etudiants sans poids'], autopct='%1.1f%%')
plt.title(f'pourcentage d\'Etudiants qui ont remplis leur poids')
plt.show()
```
